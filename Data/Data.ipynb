```python
import requests
import pandas as pd
```

```python
data_l = []

for i in range(int(35387/1000)+1):
    print(i)
    js = requests.get('https://data.cityofnewyork.us/resource/ji82-xba5.json?$offset='+str(i*1000))
    data_l.append(pd.DataFrame(js.json()))
```

```python
df = pd.concat(data_l)
```

```python
df.to_csv('public_fac.csv')
```

```python
data_l = []

for i in range(int(27312/1000)+1):
    print(i)
    js = requests.get('https://data.cityofnewyork.us/resource/8vgb-zm6e.json?$offset='+str(i*1000))
    data_l.append(pd.DataFrame(js.json()))

df = pd.concat(data_l)
df
#df.to_csv('property_values.csv')
```

```python
df['city_state_zip'] = df['city_state_zip'].astype(str)
df_filter = df[df['city_state_zip'].str.contains('NY')]
df_filter.loc[:, 'zip'] = df_filter['city_state_zip'].str.extract(r'(\d+)-')
```

```python
df_filter.to_csv('prop_values.csv')
```
